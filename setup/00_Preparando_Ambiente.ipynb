# Preparação do Ambiente — Arquitetura Medalhão (Olist)

Neste notebook, vamos preparar toda a estrutura do Data Lakehouse no Databricks para receber os dados do e-commerce Olist.

A ideia é criar o catálogo, os schemas de cada camada (bronze, silver, gold) e o volume de landing onde os CSVs serão armazenados antes da ingestão.

---

**Estrutura que será criada:**

```
medalhao/                          <- Catálogo
├── default/
│   └── landing/                   <- Volume (CSVs do Olist)
├── bronze/                        <- Dados brutos
├── silver/                        <- Dados tratados
└── gold/                          <- Dados analíticos / Data Marts
```

**Arquivos que serão carregados no volume landing:**

| Arquivo CSV | Tabela Bronze |
|:--|:--|
| `olist_customers_dataset.csv` | `bronze.tb_customers` |
| `olist_geolocation_dataset.csv` | `bronze.tb_geolocalizacao` |
| `olist_order_items_dataset.csv` | `bronze.tb_order_items` |
| `olist_order_payments_dataset.csv` | `bronze.tb_order_payments` |
| `olist_order_reviews_dataset.csv` | `bronze.tb_order_reviews` |
| `olist_orders_dataset.csv` | `bronze.tb_orders` |
| `olist_products_dataset.csv` | `bronze.tb_products` |
| `olist_sellers_dataset.csv` | `bronze.tb_sellers` |
| `product_category_name_translation.csv` | `bronze.tb_product_category_name_translation` |

Além desses, a tabela `bronze.tb_cotacao_dolar` será populada via API do Banco Central.

### Reset do ambiente
Caso precise recriar tudo do zero, esse comando apaga o catálogo inteiro com todos os schemas e tabelas. Só executar se for necessário.

````python
%sql

DROP CATALOG IF EXISTS medalhao CASCADE;
````

### 1 - Criando o Catálogo

````python
%sql

CREATE CATALOG IF NOT EXISTS medalhao;
````

### 2 - Criando os Schemas (um para cada camada)

````python
%sql

USE CATALOG medalhao;

CREATE SCHEMA IF NOT EXISTS bronze;
CREATE SCHEMA IF NOT EXISTS silver;
CREATE SCHEMA IF NOT EXISTS gold;
````

### 3 - Criando o Volume de Landing
É aqui que os CSVs do Olist vão ficar armazenados antes de serem ingeridos na camada Bronze. O caminho será `/Volumes/medalhao/default/landing/`.


````python
%sql

USE SCHEMA default;
CREATE VOLUME IF NOT EXISTS landing;
````

### 4 - Validação
Conferindo se tudo foi criado corretamente.

````python
%sql

SHOW SCHEMAS IN medalhao;
````

````python
%sql

SHOW VOLUMES IN medalhao.default;
````
